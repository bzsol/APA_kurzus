A programozásban vannak megengedett műveleteket, amiket lehetséges bármilyen programnyelv esetén használni.
A legtöbb esetben az értékadás és az egyenlőség vizsgálat szokott legtöbbször összekeveredni.

```python
# Mint látható az értékadás egy egyenlőség jel felhasználásával történik
x = 10
y = 10
# Lehetőség van használni az aritmetikai műveleteket mint a + - / * és nagyon fontos a maradékos osztást % (modulo)
z = int(x + y - (((y - x) / 1) % 2)) # int() -> egész szám legyen Python specifikus
# Az egyenlőség vizsgálat pedig két egyenlőség jellel.
print(f"{x} és {y} az egyenlő? -> {x == y}")
# Lehetőségünk van tagadni ezt az állítást ez esetben tudunk nem egyenlőséget is vizsgálni
print(f"{x} és {y} az nem egyenlő? -> {x != y}")
# Használhatók relációk is hogy a feltételeinket vizsgáljuk mint < vagy > akár <= vagy >=
print(f"Az {x} nagyobb mint {z}? -> {x > z}")
print(f"Az {x} kisebb mint {z}? -> {x < z}")
print(f"Az {x} kisebb egyenlő mint {y}? -> {x <= y}")
print(f"Az {x} nagyobb egyenlő mint {z}? -> {x >= z}")
# Lehetőségünk van tagadni is az állításainkat not segítségével vagy más nyelvek esetén a felkiátójellel -> !
print(f"{x} és {y} az egyenlő? -> {x == y}")
print(f"{x} és {y} az NEM egyenlő? -> {not x == y}")
```

A logikai műveletek közül párat jó ha ismer az ember fejből a legfontosabbak a képzés során az AND az OR és a XOR lesznek, a tagadás azaz a NOT is fontos.

### Az alapkövei a programozásnak: Szekvencia, Elágazás, Ciklusok

A szekvencia mint látható volt nem igazán igényel sok magyarázatot, amint elkeződik programunk az első sorban és haladunk szépen lefelé míg az utolsó sor végéhez érve bezár a programunk.

```python
x = 10
y = 20
print(x + y)
```

Másik nagyon fontos alapkő az elágazás. Életünk során is sokszor döntenünk kell így a programozásban is felvetődik hogy el kell döntenünk bizonyos feltételeket és az ebből kapott eredményből szeretnénk további lépéseket megoldani.
a felépítése egy if-else alapján történik, azaz ha a feltételem igaz akkor belelépek az adott feltétel részbe és ha ez nem teljesül akkor az else ágba jutok.


```python
x = 10
y = 20
if x > y:
    print(f"{x} nagyobb mint {y}")
else:
    print(f"{y} nagyobb mint {x}")
```

A szekvenciák esetén a legfontosabb tudnivaló az hogy a feltétel **amíg igaz** addig folyamatosan ismételjük a ciklusmagban megírt kódunkat.

```python
i = 1
while(i <= 10):
    print(i,end=' ')
    i = i + 1
```
